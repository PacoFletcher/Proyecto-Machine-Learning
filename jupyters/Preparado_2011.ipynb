```python
df1 = df1.rename(columns={
    "Número d'expedient": "Numero_expedient",
    "Codi districte": "Codi_districte",
    "Nom districte": "Nom_districte",
    "NK barri": "Codi_barri",
    "Nom barri": "Nom_barri",
    "Codi carrer": "Codi_carrer",
    "Nom carrer": "Nom_carrer",
    "Num postal caption": "Num_postal ",
    "Descripció dia setmana": "Descripcio_dia_setmana",
    "NK Any": "NK_Any",
    "Mes de any": "Mes_any",
    "Nom mes": "Nom_mes",
    "Dia de mes": "Dia_mes",
    "Hora de dia": "Hora_dia",
    "Descripció torn": "Descripcio_torn",
    "Descripció causa vianant": "Descripcio_causa_vianant",
    "Número de morts": "Numero_morts",
    "Número de lesionats lleus": "Numero_lesionats_lleus",
    "Número de lesionats greus": "Numero_lesionats_greus",
    "Número de víctimes": "Numero_victimes",
    "Número de vehicles implicats": "Numero_vehicles_implicats",
    "Coordenada UTM (Y)": "Coordenada_UTM_Y_ED50",
    "Coordenada UTM (X)": "Coordenada_UTM_X_ED50"
})
```

```python
df2 = df2.rename(columns={
    "Número d'expedient": "Numero_expedient",
    "Codi districte": "Codi_districte",
    "Nom districte": "Nom_districte",
    "Codi barri": "Codi_barri",
    "Nom barri": "Nom_barri",
    "Codi carrer": "Codi_carrer",
    "Nom carrer": "Nom_carrer",
    "Num postal caption": "Num_postal ",
    "Descripció dia setmana": "Descripcio_dia_setmana",
    'Dia setmana':"Dia_setmana",
    "NK Any":"NK_Any",
    "Mes de any": "Mes_any",
    "Nom mes": "Nom_mes",
    "Dia de mes": "Dia_mes",
    "Hora de dia": "Hora_dia",
    "Descripció torn": "Descripcio_torn",
    "Descripció tipus accident": "Descripcio_tipus_accident",
    "Coordenada UTM (Y)": "Coordenada_UTM_Y_ED50",
    "Coordenada UTM (X)": "Coordenada_UTM_X_ED50"
})
```

```python
df = df1.merge(df2[["Numero_expedient","Descripcio_tipus_accident"]], how = "inner", on="Numero_expedient")
```

```python
df
```

```python
df[df["Numero_expedient"].isin( df[df["Numero_expedient"].duplicated()]["Numero_expedient"])].sort_values("Numero_expedient")
```

```python
df.rename(columns={"Tipus_accident":"Descripcio_tipus_accident"},inplace=True)
```

```python
df = df.rename(columns={
    'Coordenada_UTM_X_ED50':"Coordenada_UTM_X",
       'Coordenada_UTM_Y_ED50':"Coordenada_UTM_Y",
         'Longitud_WGS84':"Longitud",
           'Latitud_WGS84':"Latitud"
})
df
```

```python
df["Codi_barri"] = df["Codi_barri"].astype(str).str.extract(r'-?(\d+)$')
df
```

```python
df["Longitud"] = None
df["Latitud"]= None
```

```python
df.to_csv("../data/df_2011.csv", index = False)
```
