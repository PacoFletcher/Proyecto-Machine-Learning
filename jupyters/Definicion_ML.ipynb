### OBJETIVOS

Define aquí el objetivo del proyecto

En el siguiente proyecto se pretende analizar la relación entre las diferentes causas de un accidente de tráfico y el número de personas implicadas con su siniestralidad.

En el conjunto de datos se recoge información como:
* Fecha y hora 
* Distrito, Barrio y Calle
* Coordenadas donde ha sucedido
* Número de víctimas
* Número de coches implicados
* Causa del peatón (en caso de que sea el causante)
* Descripción del accidente
* Número de heridos leves, graves y fallecidos

Partiendo de esta información se buscan dos principales objetivos:

1. Realizar un análisis de puntos calientes de accidentes de tráfico, así como un sistema de clasificación no supervisado que agrupe en función de ubicación y siniestralidad.
2. Un algoritmo de clasificación que, en función de las diferentes características (excluyendo heridos y fallecidos) determine un grado de siniestralidad para dar alerta al servicio de emergencias.
