```python
df1 = pd.read_csv("../data/2010_ACCIDENTS_GU_BCN_2010.csv", encoding="Windows 1252")
df2 = pd.read_csv("../data/2010_ACCIDENTS_TIPUS_GU_BCN_2010.csv", encoding="cp850")
```

```python
col1_corr = df1.columns.str.replace(" ","_")
col1_corr = col1_corr.str.replace("_de","")
col1_corr=col1_corr.str.replace("d'","")
col1_corr=col1_corr.str.replace("ú","u")
col1_corr
```

```python
dict = {}
for i,val in enumerate(df1.columns):
    dict[val] = col1_corr[i]

dict
```

```python
df1.rename(columns=dict,inplace=True)
df1
```

```python
df1.rename(columns={"Número d expedient":"Numero_expedient"}, inplace=True)
```

```python
df1.rename(columns={"Número d'expedient":"Numero_expedient"}, inplace=True)
```

```python
df = df1.merge(df2[["Numero_expedient","Descripcio_tipus_accident"]], how = "inner", on="Numero_expedient")
```

```python
df[df["Numero_expedient"].isin( df[df["Numero_expedient"].duplicated()]["Numero_expedient"])].sort_values("Numero_expedient")
```

```python
df.rename(columns={"Tipus_accident":"Descripcio_tipus_accident"},inplace=True)
```

```python
dum = pd.get_dummies(df["Descripcio_tipus_accident"]).astype("uint8")
```

```python
df = pd.concat([df.drop(columns="Descripcio_tipus_accident"),dum],axis = 1)
df
```
