```python
import pyproj
```

```python
utm_zone = pyproj.CRS("epsg:25831")
```

```python
def utm_to_latlon(x, y):
    transformer = pyproj.Transformer.from_crs(utm_zone, "epsg:4326", always_xy=True)
    lon, lat = transformer.transform(x, y)
    return lat, lon
```

Observamos que para todas las entradas no se ha guardado correctamente el decimal, por lo que dividimos entre 1000. 

Se perderá precisión, pero el error longitudinal supone unos 110 metros aproximadamente y el latitudinal unos 85 metros.

Longitud circunferencia de la Tierra en el paralelo $40 \approx 30600$ km. 

$$\frac{30600}{360\cdot1000} \approx 0,085$$

Longitud circunferencia de la Tierra para cualquier meridiano $\approx 40008$ km. 

$$\frac{40008}{360\cdot1000} \approx 0,111$$


**Apunte:** Para hacernos una idea de la significancia de este error, si lo comparamos con las manzanas del Eixample, estas tienen una longitud de unos 113 metros, por lo que las coordenadas obtenidas nos mostrarían un punto en una distancia de una calle (distancia Manhattan) en este barrio. 



```python
df["Longitud"] = np.where(df["Longitud"] > 180, df["Longitud"]/1000, df["Longitud"])
df["Latitud"] = np.where(df["Latitud"] > 180, df["Latitud"]/1000, df["Latitud"])
```

```python
from datetime import datetime, date
import holidays

# Define el rango de fechas
fecha_inicio = datetime(2010, 1, 1)
fecha_fin = datetime(2025, 12, 31)

# Crea una serie de fechas
dias = pd.date_range(fecha_inicio, fecha_fin)

# Obtiene la lista de festivos de Barcelona
festivos = holidays.CountryHoliday('ES', "CT",years=(i for i in range(2010,2025)))

# Filtra la serie para eliminar los fines de semana y festivos
dias_laborables = dias[~(dias.isin(festivos.keys())) & (dias.weekday < 5)]

# Convierte la serie a una lista de objetos datetime
lista_dias_laborables = dias_laborables.to_list()
# Imprime la lista de días laborables
print(lista_dias_laborables)
#print(festivos)
```
