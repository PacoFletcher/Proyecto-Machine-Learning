```python
import pyproj
```

```python
utm_zone = pyproj.CRS("epsg:25831")
```

```python
def utm_to_latlon(x, y):
    transformer = pyproj.Transformer.from_crs(utm_zone, "epsg:4326", always_xy=True)
    lon, lat = transformer.transform(x, y)
    return lat, lon
```

```python
df["Coordenada_UTM_X"].isna().value_counts()
```

```python
df["Coordenada_UTM_X"].describe()
```

```python
lat2, lon2 = utm_to_latlon(df["Coordenada_UTM_X"], df["Coordenada_UTM_Y"])
```

```python
df["Latitud"] = df["Latitud"].fillna(pd.Series(lat2))
df["Longitud"] = df["Longitud"].fillna(pd.Series(lon2))
```
